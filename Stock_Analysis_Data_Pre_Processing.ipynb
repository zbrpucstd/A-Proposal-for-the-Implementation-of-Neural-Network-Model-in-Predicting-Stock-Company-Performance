```python
import pandas as pd
import numpy as np
```

```python
import os
from sklearn.utils import shuffle
# Specify the folder where the files are located
folder = '/content/drive/MyDrive/NNFL-Lab/NIFTY-50 Stock/'

# Create an empty list to store the DataFrames
df_list = []

# Use a for loop to iterate through the files in the folder
for file in os.listdir(folder):
    if file.endswith('.csv'):
        # Read the file into a DataFrame
        df = pd.read_csv(os.path.join(folder, file))
        # Append the DataFrame to the list
        df_list.append(df)

# Concatenate the DataFrames into one
merged_df = pd.concat(df_list)

# Save the merged DataFrame to a new CSV file
merged_df.to_csv('merged_file.csv', index=False)

# shuffle the data
# merged_df = shuffle(merged_df)
```

```python
# You can use the pandas library to add a column with random data to a DataFrame. You can use the numpy library to generate random data and then assign it to a new column in the DataFrame
random_data = np.random.randint(11, 1643015, size=len(merged_df))
merged_df['Trades'] = random_data

random_data = np.random.randint(5, 232530747, size=len(merged_df))
merged_df['Deliverable Volume'] = random_data

random_data = np.random.uniform(0.0236, 1, size=len(merged_df))
merged_df['%Deliverble'] = random_data

# Please note that the drop() method returns a new DataFrame with the column removed and the original DataFrame remains unchanged, unless you re-assign the dropped DataFrame to the same variable name.
merged_df = merged_df.drop(['Date','Series'], axis=1)

print(merged_df)
```

```python
# For one hot encoding
merged_df = pd.get_dummies(merged_df, columns=['Symbol'])
```

```python
merged_df
```

```python
# For label binarizer
from sklearn.preprocessing import LabelBinarizer

# Create an instance of the LabelBinarizer class
lb = LabelBinarizer()

# Binarize the 'Symbol' column of the DataFrame
binarized = lb.fit_transform(merged_df['Symbol'])

# Print the binarized data
# print(binarized)

# By default, LabelBinarizer returns a dense numpy array, if you want to append it to your dataframe
# You have to convert it to dataframe and append it to your original dataframe
binarized_df = pd.DataFrame(binarized, columns=lb.classes_)
# print(binarized_df)

# Append the binarized data as a new column to the original DataFrame
merged_df = merged_df.reset_index(drop=True)
binarized_df = binarized_df.reset_index(drop=True)
merged_df = pd.concat([merged_df, binarized_df], axis=1)

# Remove the previous column
merged_df = merged_df.drop(columns=['Symbol'])
```

```python
from sklearn.preprocessing import MinMaxScaler

# Initialize the scaler
scaler = MinMaxScaler()

# Fit and transform the data
merged_df = scaler.fit_transform(merged_df)
```
